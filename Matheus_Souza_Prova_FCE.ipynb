#### Faça as seguintes comparações:

####(a) Linguagens compiladas vs. interpretadas, destacando o processo de execução.

####(b) Programação imperativa vs. procedural, relacionando com eficiência e organização.

**R:**

(a) As linguagens compiladas, são linguagens tipo o C, onde é necessário um compilador, que vai juntar todo o código e executar ele de uma única vez. Enquanto, as linguangens interpretadas, como o phython, são linguagens que executam o código linha por linha até executar todo o programa.

Ademais, as linguagens compiladas tem um execução maiz rápida do código, emitindo o resultado final em um menor tempo. Já as liguagens interpretadas tem uma maior facilidade de correção do código, pois como ela executa cada linha, ela exibe caso tenha algum erro e para a execução do programa ali, até ser corrigido o erro.

**R:**

(b) A programação imperativa é utilizada normalmente em problemas mais simples, onde não é necessário muitas linhas de código, pois a linguagem imperativa é toda a escrita do código em um único arquivo, contendo todo o programa. Enquanto isso, a linguagem procedural, utiliza a linguagem iperativa, mas foca em separar o código em módulos(arquivos diferentes), cada um com sua respectiva função, separados e unidos no código principal.

Além disso, a linguagem procedural tem uma maior organizão, principalmente para projetos grandes, pois cada etapa do código está separada, facilidando a visualização do programa, alterações, correções ou acréssimos futuros. Já a programação imperativa, mesmo tendo a organização dentro do código, em projetos grandes não é viável devido a dificuldade de encontrar as partes do código.

```python
#Funções
import numpy as np

def GerarPressão(PressMin = 1, PressMax = 6):
    '''
    Gera um número aleatório de pressão entre 1 e 6 (pressão miníma e máxima)

    Parametros:
    PressMin: int, pressão mínima
        Recebe o menor valor de pressão (bar) do intervalo. Neste caso = 1
    PressMax : int, pressão máxima
        Recebe o maior valor de pressão (bar) do intervalo. Neste caso = 6

    Returna:
    P : float
        Valor aleatório de pressão (bar) gerado

    '''
    P = np.random.rand()*(PressMax - PressMin) + PressMin #Gera 1 valor aleatório entre 0 e PressMax - PressMin (entre 0 e 5), o "+ PressMin" desloca o intervalo para o intervalo desejado (entre 1 e 6), por padão a função sempre começa o intervalo pelo 0.
    return P

def VerificarPressão(PressGer ,PressMin = 2, PressMax = 5):
    '''
    Verifica se a pressão gerada está dentro do intervalo ideal (2, 5)

    Parametros:
    PressGer: float, pressão (bar) gerada aleatoriamente
    PressMin: int, pressão mínima
        Recebe o menor valor de pressão (bar) ideal do intervalo. Neste caso = 2
    PressMax : int, pressão máxima
        Recebe o maior valor de pressão (bar) ideal do intervalo. Neste caso = 5

    Returna:
    True: se a pressão estiver dentro do intervalo seguro
    False: se estiver fora do intervalo seguro

    '''
    try: #try/except para retornar um erro caso entre algo que não seja número
        if not isinstance(PressGer, (int, float)):
            raise TypeError("Digite um Número")

        if PressMin <= PressGer <= PressMax: #Verifica se a pressão está dentro do intervalo ideal
            return True
        else:
            return False

    except TypeError as e: #caso o código seja reutilizado, essa parte pode ser importante
        print(f"erro : {e}")
```
